# 문제 3

```
import pandas as pd

# 1. 데이터 로드
url = "https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch3/members.csv"
df = pd.read_csv(url)

# 2. 전처리: 결측치 제거 후 상위 70% 데이터 추출
df = df.dropna().iloc[:int(len(df.dropna()) * 0.7)]

# 3. f1 컬럼의 1사분위수 계산 (정수 변환)
result = int(df['f1'].quantile(0.25))

print(result)
```

# 문제 4

```
import pandas as pd

# 1. 데이터 로드
url = "https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch3/members.csv"
df = pd.read_csv(url)

# 2. 결측치 확인 -> 내림차순 정렬 -> 최상단 인덱스 추출
result = df.isnull().sum().sort_values(ascending=False).index[0]

print(result)
```

# 문제 5

```
import pandas as pd

# 1. 데이터 로드
url = "https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch4/data4-1.csv"
df = pd.read_csv(url)

# 2. 3사분위수와 1사분위수의 차이 계산
# 사실 q3가 항상 크거나 같기 때문에 abs() 없이 (q3 - q1)로 작성해도 무방합니다.
diff = df['age'].quantile(0.75) - df['age'].quantile(0.25)

# 3. 소수점 이하 버림 및 정수 출력
print(int(diff))
```

# 문제 6

```
import pandas as pd

# 1. 데이터 로드
url = "https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch4/data4-2.csv"
df = pd.read_csv(url)

# 2. 반복되는 비율 계산을 변수로 선언 (가독성 향상)
ratio = (df['loves'] + df['wows']) / df['reactions']

# 3. 조건 설정
cond_ratio = (ratio > 0.4) & (ratio < 0.5)
cond_type = (df['type'] == 'video')

# 4. 결과 출력 (필터링 후 행의 개수 구하기)
result = len(df[cond_ratio & cond_type])
print(result)
```

# 문제 7

```
import pandas as pd

# 1. 데이터 로드
url = "https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch4/data4-2.csv"
df = pd.read_csv(url)

# 2. 비율 계산 (파생 변수로 생성하면 필터링 시 편리함)
# 열 계산 시 가독성을 위해 별도 변수로 추출
ratio = (df['loves'] + df['wows']) / df['reactions']

# 3. 다중 조건 설정 (괄호와 비트 연산자 & 사용 필수)
# .between()을 사용하면 0.4 < x < 0.5를 더 직관적으로 표현 가능
cond_ratio = ratio.between(0.4, 0.5, inclusive='neither')
cond_type = (df['type'] == 'video')

# 4. 결과 출력 (True의 개수 합산이 len()보다 직관적이고 빠름)
result = (cond_ratio & cond_type).sum()

print(result)
```

# 문제 8

```
import pandas as pd

# 1. 데이터 로드
url = "https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch4/data4-2.csv"
df = pd.read_csv(url)

# 2. 파생 변수(비율) 계산
ratio = (df['loves'] + df['wows']) / df['reactions']

# 3. 조건 생성 (between 활용으로 가독성 극대화)
cond_ratio = ratio.between(0.4, 0.5, inclusive='neither') # 0.4 < x < 0.5
cond_type = (df['type'] == 'video')

# 4. 결과 출력 (True의 개수를 합산)
result = (cond_ratio & cond_type).sum()
print(result)
```
