# 2주차 문제풀이 type 1

```
import pandas as pd

# 1. 데이터 로드
url = "https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch2/members.csv"
df = pd.read_csv(url)

# 2. 데이터 슬라이싱 (상위 80%)
# int()를 활용해 인덱스 번호를 계산합니다.
split_point = int(len(df) * 0.8)
df_subset = df.iloc[:split_point].copy() # 원본 보존을 위해 .copy() 권장

# 3. 결측치 처리 전 표준편차 계산
std_before = df_subset['f1'].std()

# 4. 결측치 채우기 (중앙값 활용)
f1_median = df_subset['f1'].median()
df_subset['f1'] = df_subset['f1'].fillna(f1_median)

# 5. 결측치 처리 후 표준편차 계산
std_after = df_subset['f1'].std()

# 6. 두 표준편차 차이의 절대값 계산 및 반올림
result = abs(std_before - std_after)
print(round(result, 2))
```

# 2주차 문제풀이 type 2

```
import pandas as pd

# 1. 데이터 로드
url = "https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch2/members.csv"
df = pd.read_csv(url)

# 2. 기초 통계량 계산 (표준편차, 평균)
std_val = df['age'].std()
mean_val = df['age'].mean()

# 3. 이상치 경계값(Threshold) 설정
# 기준: 평균으로부터 ±1.5 표준편차
lower_bound = mean_val - (std_val * 1.5)
upper_bound = mean_val + (std_val * 1.5)

# 4. 이상치 필터링 조건 설정
# 'age'가 하한값보다 작거나 상한값보다 큰 경우
is_outlier = (df['age'] < lower_bound) | (df['age'] > upper_bound)

# 5. 결과 출력 (이상치들의 'age' 합계)
outlier_sum = df.loc[is_outlier, 'age'].sum()
print(outlier_sum)
```
